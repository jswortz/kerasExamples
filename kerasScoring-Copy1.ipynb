```python
# JSON Scoring Function 

import pickle
from keras.models import load_model
import numpy as np
import json
# import pandas as pd


######data preprocessing
with open('kerasTransformers.pkl', 'rb') as f:
    u = pickle._Unpickler(f)
    u.encoding = 'latin1'
    transformers = u.load()
    
categorTrans = transformers[1]
numericTrans = transformers[0]


#load the model
# loaded_model = load_model('model2.hd5')  
with open('sklearnSVCVanilla.pkl', 'rb') as f:
    u = pickle._Unpickler(f)
    u.encoding = 'latin1'
    loaded_model = u.load()
    
def flat2gen(alist):
    for item in alist:
        if isinstance(item, list):
            for subitem in item: yield subitem
        else:
            yield item

def catScore(labelEnc, oneHot, column, data):
#     labeled = labelEnc.transform(data[column].apply(str))
    labeled = labelEnc.transform([str(data[column])])
    labeled = [[x] for x in labeled]
    return oneHot.transform(labeled)

def contScore(stScale, column, data):
#     tmp = data[column].values.reshape(1,-1) 
    return stScale.transform([data[column]])

def scoreFun(JSONdata, model=loaded_model):
#     inputData = pd.read_json(JSONdata)
    inputData = json.loads(JSONdata)[0]
    contTransformed = [contScore(x[1], x[0], inputData) for x in numericTrans]
    catTransformed = [catScore(x[1], x[2], x[0], inputData) for x in categorTrans]
    dataFlat = []
    
    for x in contTransformed:
        dataFlat.append(x)

    for x in catTransformed:
        l = x.tolist()
        dataFlat.append(l)

    dataFlat = zip(*dataFlat)

    dataFlat = [list(x) for x in dataFlat]

    df = []
    for line in dataFlat:
        test = [x for x in flat2gen(line)]
        df.append(test)
    
    return model.predict_proba(df)

#################TESTS BELOW
testData = '[{"pendingpaymentamount": 0.0, "anilearning": "NOT_FOUND", "broadcastfunction": "Array", "currentbalance": 203.5, "install": "false", "previousdueamount": 0.0, "projectedbalancedue": 160.08, "previouspaymentamount": 160.04, "daysdelinquent": 0, "existingrecurringpaymentonaccount": false, "cancelledappt": "false", "statmentbalance": 0.0, "outageflag": "false", "recentpaymentamount": 160.04, "rec_pay_diff": -9999, "accountstatus": "ACTIVE", "nonpaydiscoii": 0, "promise_diff": -9999, "statementbalance": 0.0, "balancedue": 160.08, "pastduebalance": 0.0, "delinq_days": 0, "cardpaymentrestricted": false, "boradcaststatus": "NOT_FOUND", "bankpaymentrestricted": false, "prev_pay_diff": -31, "past_due_diff": -9999, "due_diff": 7, "bal_due_diff": 7, "billing": 1}]'
scoreFun(testData)
    
```
